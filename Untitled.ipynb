```python
import pandas as pd

# Load data
df = pd.read_csv('user_data.csv')
df.head()
```

```python
import matplotlib.pyplot as plt

# Plot income by age
age_income = df.groupby('Age')['Income'].sum().sort_values(ascending=False)
age_income.plot(kind='bar', color='skyblue')
plt.title('Ages with the Highest Income')
plt.xlabel('Age')
plt.ylabel('Total Income')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file into a DataFrame
df = pd.read_csv('user_data.csv')

# Show the first few rows of the data
df.head()

# Now you can use 'df' for further processing or plotting
# Example: Show Ages with the Highest Income
age_income = df.groupby('Age')['Income'].sum().sort_values(ascending=False)
age_income.plot(kind='bar', color='skyblue')
plt.title('Ages with the Highest Income')
plt.xlabel('Age')
plt.ylabel('Total Income')
plt.show()

# Example: Gender Distribution Across Spending Categories
df['Spending_Category'] = pd.cut(df['Spending'], bins=[0, 1000, 3000, 5000], labels=['Low', 'Medium', 'High'])
sns.countplot(data=df, x='Spending_Category', hue='Gender', palette='pastel')
plt.title('Gender Distribution Across Spending Categories')
plt.xlabel('Spending Category')
plt.ylabel('Count')
plt.show()
```

```python
plt.savefig('age_income_chart.png')  # Save the income plot
plt.savefig('gender_distribution_chart.png')  # Save the gender distribution chart
```
